```python
import time
# Initialising the Data file

fd = open('Inventory.txt','r')
txt = fd.read()
fd.close()
items = txt.split("\n");
# print(txt)

# Taking user input
Customer_name = input("Enter your name : ")
Customer_phone_no = input("Enter your phone number : ")
Customer_email_id = input("Enter your email id : ")
Product_id = input("Enter the required product id : ")
qnt = input("Enter the required quantity :")
print("-"*20)

updated_lst=[]


for product in items:
    # Splitting the product details
    ls = product.split(',')

    # Checking the required product id
    if ls[0]==Product_id:

        print("Required Product ID:",Product_id)
        print("-"*20)
        print("Required Product :" , ls[1])
        print("Required Quantity is :" , qnt)
        print("-"*20)

        # Checking the availability of the product
        if int(ls[3])>= int(qnt):
            # Calculating the billing amount
            amt = int(qnt)*int(ls[2])
            print("Billing Amount :" , amt)
            # Updating the quantity of the product
            ls[3] = str(int(ls[3])-int(qnt))
            print("ThankYou Visit Again !!")

            # Updating the customer details in the customer.txt file
            fs = open('customer.txt' , 'a')
            pr = "Customer Name: " + Customer_name + '\n' + "Phone No. : " + Customer_phone_no + '\n' + "Customer_email : "  + Customer_email_id + '\n' + "Product_Id_required : " + Product_id + '\n' + "Quantity of the Product : " + qnt + '\n' + "Total Bill Amount : " +  str(amt) + '\n' + "Product Purchase Timings : " +  time.ctime() +'\n' + '-'*20 + '\n'
            fs.write(pr)
            fs.close()


        else :

            if int(ls[3])==0:
                print("We are Sorry to Inform You that we are out of stock of the product you require.")
            else :
                print("We are Sorry to Inform You that we are in short of the product you require.")
                print("However we can serve the remaining quantity of the product you require.")
                print("-"*20)

                inp = input("Do you want to continue (Y/N) :")

                if inp=='Y' or inp=='y':
                    amt = int(ls[3])*int(ls[2])
                    print("As per your request we can serve you with the remaining quantity of the product you require.")
                    print("Billing Amount :" , amt)
            
                    # Updating the customer details in the customer.txt file
                    fs = open('customer.txt' , 'a')
                    pr = "Customer Name: " + Customer_name + '\n' + "Phone No. : " + Customer_phone_no + '\n' + "Customer_email : "  + Customer_email_id + '\n' + "Product_Id_required : " + Product_id + '\n' + "Quantity of the Product : " + ls[3] + '\n' + "Total Bill Amount : " +  str(amt) + '\n' + "Product Purchase Timings : " +  time.ctime() +'\n' + '-'*20 + '\n'
                    fs.write(pr)
                    fs.close()

                    ls[3] = '0'
                    print("ThankYou Visit Again !!")
                else:
                    print("We are Sorry for the inconvenience caused.")
                    print("ThankYou Visit Again !!")
                    print("-"*20)
            
    # Updating my inventory list
    updated_lst.append(ls)

# removed the last element of the list as it is empty
updated_lst.pop()

# Updating the Inventory.txt file
fd = open('Inventory.txt' , 'w')
for u in updated_lst:
    prod = u[0] + ',' + u[1] + ',' + u[2] + ',' + u[3] + '\n'
    fd.write(prod)
fd.close()

fd = open('Inventory.txt','r')
txt = fd.read()

print("Updated the Inventory :")
print(txt)

```
