```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Position_Salaries.csv')
```

```python
df.head(10)
```

## `Splitting the data`

```python
x = df.iloc[:,1:-1].values 
y = df.iloc[:,-1].values
```

### **Importing the Linear Regression model library**

```python
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
```

### **Training the model with input and output**

```python
lin_reg.fit(x,y)
```

```python
pred_val = lin_reg.predict(x)
```

### `Visualising the Simple Linear Regression Model`

```python
plt.scatter(x,y,color='red')
plt.plot(x,pred_val,color='blue')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.legend(['True Value' , 'Predicted Value'])
plt.show()
```

## `Polynomial Linear Regression`

### **Implementing Polynomial Features**

```python
from sklearn.preprocessing import PolynomialFeatures

poly_reg = PolynomialFeatures(degree=4)

x_poly = poly_reg.fit_transform(x)
```

```python
poly_lin_reg = LinearRegression()

poly_lin_reg.fit(x_poly,y)
```

```python
poly_pred_val = poly_lin_reg.predict(x_poly)
```

### `Visalising the Polynomial Regression Model with degree 2`

```python
plt.scatter(x,y,color='red')
plt.plot(x,poly_pred_val,color='blue')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.legend(['True Value' , 'Predicted Value'])
plt.show()
```

### `Visalising the Polynomial Regression Model with degree 3`

```python
plt.scatter(x,y,color='red')
plt.plot(x,poly_pred_val,color='blue')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.legend(['True Value' , 'Predicted Value'])
plt.show()
```
